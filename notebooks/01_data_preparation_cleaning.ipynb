# EDA y preparación de datos: inversión turística en Málaga ciudad

## Contexto del proyecto

Una inversora privada quiere comprar pisos en Málaga ciudad para destinarlos a vivienda turística. El objetivo de este análisis es explorar un dataset de alojamientos turísticos y preparar los datos necesarios para construir un dashboard interactivo en Power BI.

El dashboard permitirá comparar distritos de Málaga según precio por noche, ocupación estimada, ingresos estimados, reviews, tipo de alojamiento y nivel de competencia.

## Pregunta principal

¿Qué distritos de Málaga ciudad presentan mejores oportunidades para estudiar una inversión en vivienda turística?

## Alcance del análisis

Este análisis no determina la rentabilidad final de una inversión inmobiliaria, ya que el dataset no incluye precio de compra de viviendas, impuestos, reformas, hipoteca, comunidad, gastos de limpieza ni normativa urbanística.

El objetivo es realizar una primera criba de zonas con señales interesantes dentro del mercado turístico.

```python
# ============================================================
# 1. CONFIGURACIÓN INICIAL DEL PROYECTO
# ============================================================
# En esta celda importamos las librerías principales que vamos a usar
# durante el EDA y dejamos preparadas las rutas de los archivos.
#
# EDA significa "Exploratory Data Analysis", es decir, análisis exploratorio.
# Aquí vamos a revisar, entender y preparar el dataset antes de llevarlo a Power BI.
# ============================================================


# Librería principal para trabajar con datos en formato tabla
import pandas as pd

# Librería para operaciones numéricas
import numpy as np

# Librería básica para crear gráficos durante el EDA
import matplotlib.pyplot as plt

# Librería para trabajar con rutas de archivos de forma más segura
from pathlib import Path

# Librería para comprobar qué Python/kernel está usando el notebook
import sys


# ------------------------------------------------------------
# Configuración visual de pandas
# ------------------------------------------------------------
# Mostramos todas las columnas cuando visualicemos tablas.
# Esto nos ayudará porque el dataset tiene muchas columnas.
pd.set_option("display.max_columns", None)

# Limitamos el número de filas que se muestran para no saturar la pantalla.
pd.set_option("display.max_rows", 20)

# Configuramos los números decimales para que se vean con 2 decimales.
pd.set_option("display.float_format", "{:,.2f}".format)


# ------------------------------------------------------------
# Definición de rutas del proyecto
# ------------------------------------------------------------
# Ruta del dataset original, guardado en la carpeta data/raw.
ruta_dataset_raw = Path("../data/raw/listings.csv")

# Ruta donde más adelante guardaremos el dataset preparado para Power BI.
ruta_dataset_procesado = Path("../data/processed/listings_malaga_processed.csv")


# ------------------------------------------------------------
# Comprobaciones iniciales
# ------------------------------------------------------------
# Mostramos qué Python está usando el notebook.
# Debería aparecer una ruta que contenga ".venv".
print("Python utilizado por el notebook:")
print(sys.executable)

# Comprobamos si el archivo listings.csv existe en la ruta esperada.
print("\n¿Existe el archivo del dataset original?")
print(ruta_dataset_raw.exists())

# Mensaje final para confirmar que la celda se ha ejecutado bien.
print("\nConfiguración inicial completada correctamente.")
```

```python
# ============================================================
# 2. CARGA DEL DATASET ORIGINAL
# ============================================================
# En esta celda cargamos el archivo listings.csv en un DataFrame de pandas.
#
# Un DataFrame es una tabla:
# - cada fila representa un alojamiento turístico
# - cada columna representa una variable del alojamiento
#
# Usamos low_memory=False para que pandas lea el archivo con más seguridad
# cuando hay muchas columnas o tipos de datos mezclados.
# ============================================================


# Cargamos el dataset original desde la ruta definida en la celda anterior
df = pd.read_csv(ruta_dataset_raw, low_memory=False)


# ------------------------------------------------------------
# Mostramos las primeras filas del dataset
# ------------------------------------------------------------
# Esto nos permite ver rápidamente cómo vienen los datos:
# nombres de columnas, valores, formato de precios, textos, números, etc.
df.head()
```

```python
# ============================================================
# 3. DIMENSIONES DEL DATASET
# ============================================================
# Con shape comprobamos el tamaño del dataset.
#
# El resultado se interpreta así:
# (número de filas, número de columnas)
#
# En este proyecto:
# - filas = alojamientos turísticos
# - columnas = variables disponibles sobre cada alojamiento
# ============================================================


# Guardamos el número de filas y columnas en dos variables
num_filas, num_columnas = df.shape


# Mostramos el resultado de forma clara
print(f"El dataset tiene {num_filas} filas y {num_columnas} columnas.")
```

```python
# ============================================================
# 4. TABLA RESUMEN DE COLUMNAS
# ============================================================
# Creamos una tabla resumen para revisar la calidad inicial del dataset.
#
# Para cada columna veremos:
# - tipo de dato
# - número de valores informados
# - número de valores vacíos
# - porcentaje de valores vacíos
#
# Esta revisión nos ayuda a decidir qué columnas son útiles para el análisis
# y cuáles conviene descartar porque tienen demasiados valores nulos.
# ============================================================

# Creamos un DataFrame resumen con información básica de cada columna
resumen_columnas = pd.DataFrame({
    "columna": df.columns,
    "tipo_dato": df.dtypes.values,
    "valores_no_nulos": df.notnull().sum().values,
    "valores_nulos": df.isnull().sum().values
})

# Calculamos el porcentaje de valores nulos de cada columna
resumen_columnas["porcentaje_nulos"] = (
    resumen_columnas["valores_nulos"] / len(df) * 100
).round(2)

# Ordenamos de mayor a menor porcentaje de nulos
# Así vemos primero las columnas más incompletas
resumen_columnas = resumen_columnas.sort_values(
    by="porcentaje_nulos",
    ascending=False
).reset_index(drop=True)

# ------------------------------------------------------------
# Mostramos la tabla completa
# ------------------------------------------------------------
# Usamos pd.option_context para decirle a pandas que muestre
# hasta 100 filas, suficiente porque nuestro dataset tiene 79 columnas.
#
# display() muestra la tabla en formato visual dentro del notebook,
# más cómodo que print() para tablas.

with pd.option_context("display.max_rows", 100):
    display(resumen_columnas)
```

### Conclusión sobre valores nulos

El dataset original contiene columnas con un porcentaje elevado de valores nulos, como `neighbourhood_group_cleansed` o `calendar_updated`, que no aportan información útil para el objetivo del proyecto.

Sin embargo, también contiene variables con información suficiente para analizar el mercado turístico de Málaga, especialmente aquellas relacionadas con ubicación, precio, disponibilidad, reviews, ocupación estimada e ingresos estimados.

Por tanto, el dataset se considera válido para el análisis, aunque será necesario seleccionar únicamente las columnas relevantes para la pregunta de negocio.

## Selección de columnas útiles

Para responder a la pregunta de negocio, se seleccionan únicamente las variables que ayudan a comparar distritos de Málaga desde el punto de vista de una posible inversión turística.

Las columnas seleccionadas se agrupan en los siguientes bloques:

- Identificación del alojamiento.
- Ubicación y distrito.
- Tipo de alojamiento y características básicas.
- Precio y condiciones de reserva.
- Disponibilidad anual.
- Reviews como aproximación de actividad turística.
- Puntuaciones de los huéspedes.
- Ocupación estimada e ingresos estimados.

Se descartan variables como licencia, superhost o número de camas porque no se consideran criterios principales para decidir en qué zona estudiar una inversión. El análisis se centra en señales de mercado: precio, ingresos estimados, ocupación, disponibilidad, reviews, capacidad, dormitorios y competencia por distrito.

```python
# ============================================================
# 5. SELECCIÓN DE COLUMNAS ÚTILES PARA EL ANÁLISIS
# ============================================================
# El dataset original tiene muchas columnas.
# Para este proyecto no necesitamos usarlas todas.
#
# Vamos a seleccionar solo las columnas que nos ayudan a responder:
#
# ¿Qué distritos de Málaga ciudad presentan mejores señales
# para estudiar una inversión en vivienda turística?
#
# IMPORTANTE:
# No modificamos el DataFrame original "df".
# Creamos un nuevo DataFrame llamado "df_inversion".
# ============================================================


# ------------------------------------------------------------
# Lista de columnas que queremos utilizar
# ------------------------------------------------------------
# Estas columnas se han elegido porque aportan información útil
# para analizar ubicación, precio, demanda, competencia y rentabilidad estimada.

columnas_utiles = [
    # Identificación básica del alojamiento
    "id",
    "name",

    # Ubicación geográfica y distrito
    "neighbourhood_cleansed",
    "latitude",
    "longitude",

    # Tipo y características del alojamiento
    "property_type",
    "room_type",
    "accommodates",
    "bedrooms",

    # Precio y condiciones de reserva
    "price",
    "minimum_nights",

    # Disponibilidad
    "availability_365",

    # Reviews como aproximación de actividad/demanda
    "number_of_reviews",
    "number_of_reviews_ltm",
    "reviews_per_month",

    # Puntuaciones de los huéspedes
    "review_scores_rating",
    "review_scores_location",


    # Variables estimadas de rendimiento turístico
    "estimated_occupancy_l365d",
    "estimated_revenue_l365d"
]


# ------------------------------------------------------------
# Comprobamos qué columnas existen realmente en el dataset
# ------------------------------------------------------------
# Esto evita errores si alguna columna escrita arriba no existe
# exactamente con ese nombre en el CSV.

columnas_existentes = []

columnas_faltantes = []

for columna in columnas_utiles:
    if columna in df.columns:
        columnas_existentes.append(columna)
    else:
        columnas_faltantes.append(columna)


# ------------------------------------------------------------
# Mostramos el resultado de la comprobación
# ------------------------------------------------------------

print(f"Columnas solicitadas: {len(columnas_utiles)}")
print(f"Columnas encontradas en el dataset: {len(columnas_existentes)}")
print(f"Columnas no encontradas: {len(columnas_faltantes)}")

print("\nColumnas no encontradas:")
print(columnas_faltantes)


# ------------------------------------------------------------
# Creamos un nuevo DataFrame solo con las columnas útiles
# ------------------------------------------------------------
# Este será el DataFrame principal para el resto del EDA.

df_inversion = df[columnas_existentes].copy()


# ------------------------------------------------------------
# Comprobamos el tamaño del nuevo DataFrame
# ------------------------------------------------------------

print("\nTamaño del DataFrame original:")
print(df.shape)

print("\nTamaño del DataFrame seleccionado para el análisis:")
print(df_inversion.shape)


# ------------------------------------------------------------
# Mostramos las primeras filas del nuevo DataFrame
# ------------------------------------------------------------

df_inversion.head()
```

## Renombrado de columnas

Se renombran las columnas seleccionadas al español para facilitar la interpretación del análisis y del futuro dashboard en Power BI.

El dataset original se conserva sin modificar en la carpeta `data/raw`.  
El renombrado se aplica únicamente sobre el DataFrame de trabajo `df_inversion`.

```python
# ============================================================
# 6. RENOMBRADO DE COLUMNAS
# ============================================================
# Cambiamos los nombres de las columnas seleccionadas al español.
# Usamos nombres en minúsculas, sin tildes y con guion bajo.
# ============================================================


# Diccionario de renombrado:
# clave = nombre original de la columna
# valor = nuevo nombre en español
nuevos_nombres_columnas = {
    "id": "id_alojamiento",
    "name": "nombre",
    "neighbourhood_cleansed": "distrito",
    "latitude": "latitud",
    "longitude": "longitud",
    "property_type": "tipo_propiedad",
    "room_type": "tipo_alojamiento",
    "accommodates": "capacidad",
    "bedrooms": "dormitorios",
    "price": "precio",
    "minimum_nights": "noches_minimas",
    "availability_365": "disponibilidad_365",
    "number_of_reviews": "reviews_totales",
    "number_of_reviews_ltm": "reviews_ultimo_anio",
    "reviews_per_month": "reviews_por_mes",
    "review_scores_rating": "puntuacion_general",
    "review_scores_location": "puntuacion_ubicacion",
    "calculated_host_listings_count": "alojamientos_del_anfitrion",
    "estimated_occupancy_l365d": "ocupacion_estimada_365",
    "estimated_revenue_l365d": "ingresos_estimados_365"
}


# Aplicamos el renombrado al DataFrame de trabajo
df_inversion = df_inversion.rename(columns=nuevos_nombres_columnas)


# Comprobamos los nuevos nombres de las columnas
print("Columnas después del renombrado:\n")

for columna in df_inversion.columns:
    print(f"- {columna}")


# Mostramos una vista previa del DataFrame renombrado
df_inversion.head()
```

```python
# ============================================================
# 7. FILTRADO DE VIVIENDAS COMPLETAS
# ============================================================
# El objetivo del proyecto es ayudar a una inversora privada
# interesada en comprar pisos en Málaga ciudad para destinarlos
# a vivienda turística.
#
# Por ese motivo, centramos el análisis en alojamientos completos:
# "Entire home/apt".
#
# Excluimos habitaciones privadas, habitaciones compartidas u otros tipos,
# porque no representan el tipo de activo inmobiliario que la inversora
# quiere comprar.
# ============================================================


# Comprobamos cuántos alojamientos hay por tipo antes del filtrado
print("Distribución de alojamientos por tipo antes del filtrado:\n")
print(df_inversion["tipo_alojamiento"].value_counts())


# Guardamos el número de filas antes del filtrado
filas_antes_filtro = len(df_inversion)


# Filtramos solo viviendas completas
df_inversion = df_inversion[
    df_inversion["tipo_alojamiento"] == "Entire home/apt"
].copy()


# Guardamos el número de filas después del filtrado
filas_despues_filtro = len(df_inversion)


# Calculamos cuántas filas se han eliminado
filas_eliminadas_filtro = filas_antes_filtro - filas_despues_filtro


# Mostramos resumen del filtrado
print("\nFilas antes del filtrado:")
print(filas_antes_filtro)

print("\nFilas después del filtrado:")
print(filas_despues_filtro)

print("\nFilas eliminadas:")
print(filas_eliminadas_filtro)

print("\nPorcentaje de filas eliminadas:")
print(round(filas_eliminadas_filtro / filas_antes_filtro * 100, 2), "%")


# Comprobamos que ahora solo quedan viviendas completas
print("\nDistribución de alojamientos por tipo después del filtrado:\n")
print(df_inversion["tipo_alojamiento"].value_counts())
```

```python
# ============================================================
# 8. CONVERSIÓN DE LA COLUMNA PRECIO A FORMATO NUMÉRICO
# ============================================================
# La columna "precio" viene originalmente como texto.
#
# Para poder calcular medias, medianas, rankings y gráficos,
# necesitamos convertirla a formato numérico.
# ============================================================


# ------------------------------------------------------------
# Limpiamos y convertimos la columna precio
# ------------------------------------------------------------
# Pasos:
# 1. Convertimos la columna a texto por seguridad.
# 2. Quitamos el símbolo "$".
# 3. Quitamos posibles comas "," en precios grandes.
# 4. Convertimos el resultado a número.
#
# errors="coerce" convierte en NaN cualquier valor que no pueda transformarse.

df_inversion["precio"] = (
    df_inversion["precio"]
    .astype(str)
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False)
)

df_inversion["precio"] = pd.to_numeric(
    df_inversion["precio"],
    errors="coerce"
)


# ------------------------------------------------------------
# Comprobación del resultado
# ------------------------------------------------------------
# El tipo de dato debería ser float64.

print("Tipo de dato de la columna precio después de la conversión:")
print(df_inversion["precio"].dtype)


# Mostramos algunos precios ya convertidos
df_inversion[["precio"]].head(10)
```

```python
# ============================================================
# 9. MUESTRA DE COLUMNAS CLAVE
# ============================================================
# Revisamos una muestra de las variables principales después
# de seleccionar columnas, renombrarlas y convertir el precio.
#
# Esta vista nos permite comprobar que las columnas principales
# están listas para continuar con el análisis.
# ============================================================


# Columnas principales para una primera revisión visual
columnas_clave = [
    "distrito",
    "tipo_alojamiento",
    "tipo_propiedad",
    "precio",
    "capacidad",
    "dormitorios",
    "noches_minimas",
    "disponibilidad_365",
    "reviews_totales",
    "reviews_ultimo_anio",
    "reviews_por_mes",
    "puntuacion_general",
    "puntuacion_ubicacion",
    "ocupacion_estimada_365",
    "ingresos_estimados_365"
]


# Mostramos las primeras 10 filas de esas columnas
df_inversion[columnas_clave].head(10)
```

```python
# ============================================================
# 10. REVISIÓN DE VALORES NULOS EN EL DATAFRAME DE TRABAJO
# ============================================================
# En esta celda revisamos los valores nulos únicamente en las columnas
# seleccionadas para nuestro análisis.
#
# Ya no estamos revisando las 79 columnas originales, sino las variables
# que realmente vamos a usar para construir el dashboard.
#
# Esto nos ayuda a decidir:
# - qué columnas están completas
# - qué columnas tienen algunos valores vacíos
# - qué tratamiento aplicar más adelante
# ============================================================


# ------------------------------------------------------------
# Creamos una tabla resumen de nulos para df_inversion
# ------------------------------------------------------------
# Para cada columna calculamos:
# - tipo de dato
# - valores no nulos
# - valores nulos
# - porcentaje de nulos

resumen_nulos_inversion = pd.DataFrame({
    "columna": df_inversion.columns,
    "tipo_dato": df_inversion.dtypes.values,
    "valores_no_nulos": df_inversion.notnull().sum().values,
    "valores_nulos": df_inversion.isnull().sum().values
})


# ------------------------------------------------------------
# Calculamos el porcentaje de nulos
# ------------------------------------------------------------
# Dividimos los valores nulos entre el total de filas del dataset.
# Multiplicamos por 100 para obtener el porcentaje.

resumen_nulos_inversion["porcentaje_nulos"] = (
    resumen_nulos_inversion["valores_nulos"] / len(df_inversion) * 100
).round(2)


# ------------------------------------------------------------
# Ordenamos de mayor a menor porcentaje de nulos
# ------------------------------------------------------------
# Así veremos primero las columnas más incompletas.

resumen_nulos_inversion = resumen_nulos_inversion.sort_values(
    by="porcentaje_nulos",
    ascending=False
).reset_index(drop=True)


# ------------------------------------------------------------
# Mostramos el resultado
# ------------------------------------------------------------
# Usamos to_string(index=False) para que se muestren todas las filas
# sin que pandas oculte columnas intermedias con "...".

print(resumen_nulos_inversion.to_string(index=False))
```

## Tratamiento de valores nulos

Tras revisar las variables seleccionadas, se observa que la mayoría de columnas clave tienen una cobertura suficiente para el análisis.

Las columnas `precio` e `ingresos_estimados_365` presentan valores nulos en aproximadamente un 9% de los registros. Como ambas variables son fundamentales para comparar oportunidades de inversión, se eliminarán del dataset final los alojamientos que no tengan información en estas columnas.

La columna `reviews_por_mes` también presenta valores nulos, pero en este caso se rellenará con 0, interpretando la ausencia de dato como falta de actividad mensual registrada.

Las columnas `puntuacion_general` y `puntuacion_ubicacion` mantienen algunos valores nulos. No se rellenan con 0 porque eso podría interpretarse erróneamente como una mala puntuación real, cuando probablemente indica ausencia de reviews suficientes.

```python
# ============================================================
# 11. TRATAMIENTO DE VALORES NULOS
# ============================================================
# En esta celda preparamos un nuevo DataFrame para el análisis final
# y para el futuro dashboard en Power BI.
#
# Partimos de df_inversion, que ya contiene solo las columnas útiles
# para responder a la pregunta de negocio:
#
# ¿Qué distritos de Málaga ciudad presentan mejores señales
# para estudiar una inversión en vivienda turística?
#
# Decisiones principales:
#
# 1. Eliminamos registros sin precio o sin ingresos estimados.
#    Estas variables son clave para analizar oportunidad turística.
#
# 2. Rellenamos reviews_por_mes con 0.
#    Si no hay reviews mensuales informadas, lo interpretamos como
#    ausencia de actividad mensual registrada.
#
# ============================================================


# ------------------------------------------------------------
# Creamos una copia del DataFrame de trabajo
# ------------------------------------------------------------

df_dashboard = df_inversion.copy()


# ------------------------------------------------------------
# Eliminamos filas sin precio o sin ingresos estimados
# ------------------------------------------------------------

df_dashboard = df_dashboard.dropna(
    subset=["precio", "ingresos_estimados_365"]
)


# ------------------------------------------------------------
# Rellenamos reviews_por_mes
# ------------------------------------------------------------

df_dashboard["reviews_por_mes"] = df_dashboard["reviews_por_mes"].fillna(0)


# ------------------------------------------------------------
# Comprobamos cuántas filas quedan después del tratamiento
# ------------------------------------------------------------

filas_antes = len(df_inversion)
filas_despues = len(df_dashboard)
filas_eliminadas = filas_antes - filas_despues

print("Filas antes del tratamiento:")
print(filas_antes)

print("\nFilas después del tratamiento:")
print(filas_despues)

print("\nFilas eliminadas:")
print(filas_eliminadas)

print("\nPorcentaje de filas eliminadas:")
print(round(filas_eliminadas / filas_antes * 100, 2), "%")


# ------------------------------------------------------------
# Revisamos de nuevo los nulos en variables principales
# ------------------------------------------------------------

df_dashboard[
    [
        "precio",
        "ingresos_estimados_365",
        "reviews_por_mes"
    ]
].isnull().sum()
```

```python
# ============================================================
# 12. REVISIÓN FINAL DE VALORES NULOS
# ============================================================
# Comprobamos los valores nulos restantes en el dataset final
# que se usará para el análisis y el dashboard en Power BI.
# ============================================================


# Creamos una tabla resumen de nulos para el DataFrame final
resumen_nulos_dashboard = pd.DataFrame({
    "columna": df_dashboard.columns,
    "tipo_dato": df_dashboard.dtypes.values,
    "valores_no_nulos": df_dashboard.notnull().sum().values,
    "valores_nulos": df_dashboard.isnull().sum().values
})


# Calculamos el porcentaje de nulos de cada columna
resumen_nulos_dashboard["porcentaje_nulos"] = (
    resumen_nulos_dashboard["valores_nulos"] / len(df_dashboard) * 100
).round(2)


# Ordenamos de mayor a menor porcentaje de nulos
resumen_nulos_dashboard = resumen_nulos_dashboard.sort_values(
    by="porcentaje_nulos",
    ascending=False
).reset_index(drop=True)


# Mostramos la tabla final
with pd.option_context("display.max_rows", 100):
    display(resumen_nulos_dashboard)
```

### Conclusión final sobre valores nulos

Después del tratamiento, las variables principales para el análisis económico —`precio`, `ingresos_estimados_365` y `reviews_por_mes`— no presentan valores nulos.

Las columnas `puntuacion_general` y `puntuacion_ubicacion` mantienen algunos valores vacíos. Se decide conservarlos sin rellenar, ya que sustituirlos por 0 podría interpretarse erróneamente como una mala puntuación real.

La columna `dormitorios` conserva un número mínimo de valores nulos, por lo que no afecta de forma significativa al análisis. Estos casos podrán tratarse más adelante como `Sin información` si se utiliza una categoría de dormitorios en el dashboard.

El dataset final queda preparado para continuar con la creación de variables derivadas y el análisis descriptivo.

```python
# ============================================================
# 13. REVISIÓN DE VALORES EXTREMOS EN PRECIO
# ============================================================
# Antes de crear variables derivadas y métricas por distrito,
# revisamos la distribución de la columna "precio".
#
# El objetivo es detectar si existen valores extremadamente altos
# que puedan distorsionar el análisis posterior.
#
# Esta revisión es necesaria porque el precio se usará después para:
# - clasificar alojamientos por nivel de precio
# - calcular métricas medias por distrito
# - interpretar los ingresos estimados
# ============================================================


# ------------------------------------------------------------
# Resumen estadístico del precio
# ------------------------------------------------------------
# Con describe() revisamos valores como:
# - precio mínimo
# - precio medio
# - mediana
# - percentiles
# - precio máximo

print("Resumen estadístico del precio:\n")
display(df_dashboard["precio"].describe())


# ------------------------------------------------------------
# Alojamientos con precios más altos
# ------------------------------------------------------------
# Mostramos los 20 alojamientos con precio más alto
# para comprobar si existen valores anómalos o poco representativos.

print("\nAlojamientos con precios más altos:\n")

df_dashboard[
    [
        "distrito",
        "nombre",
        "tipo_propiedad",
        "precio",
        "dormitorios",
        "capacidad",
        "ingresos_estimados_365",
        "ocupacion_estimada_365"
    ]
].sort_values(
    by="precio",
    ascending=False
).head(20)
```

## Tratamiento de valores extremos en precio

Tras revisar la distribución de la columna `precio`, se detectan valores extremadamente altos que no parecen representativos del mercado habitual de viviendas turísticas en Málaga.

Estos valores pueden distorsionar métricas como el precio medio, los ingresos estimados y los rankings por distrito.

Por este motivo, se decide aplicar un filtro de precio máximo de 1.000 € por noche. Esta decisión no implica que los anuncios sean necesariamente incorrectos, sino que se consideran valores extremos para el objetivo de este análisis.

El objetivo es construir un dataset más estable y representativo para comparar distritos desde una perspectiva de inversión turística estándar.

```python
# ============================================================
# 14. TRATAMIENTO DE VALORES EXTREMOS EN PRECIO
# ============================================================
# Aplicamos un filtro para excluir precios superiores a 1.000 €.
# Estos valores se consideran extremos para el objetivo del análisis.
# ============================================================

# ------------------------------------------------------------
# Guardamos el número de filas antes del filtro
# ------------------------------------------------------------

filas_antes_outliers = len(df_dashboard)


# ------------------------------------------------------------
# Aplicamos el filtro de precio máximo
# ------------------------------------------------------------
# Nos quedamos solo con alojamientos cuyo precio sea menor o igual
# a 1.000 € por noche.

df_dashboard = df_dashboard[
    df_dashboard["precio"] <= 1000
].copy()


# ------------------------------------------------------------
# Calculamos el impacto del filtro
# ------------------------------------------------------------

filas_despues_outliers = len(df_dashboard)

filas_eliminadas_outliers = filas_antes_outliers - filas_despues_outliers

porcentaje_eliminado_outliers = (
    filas_eliminadas_outliers / filas_antes_outliers * 100
)


# ------------------------------------------------------------
# Mostramos los resultados del tratamiento
# ------------------------------------------------------------

print("Filas antes de eliminar valores extremos de precio:")
print(filas_antes_outliers)

print("\nFilas después de eliminar valores extremos de precio:")
print(filas_despues_outliers)

print("\nFilas eliminadas:")
print(filas_eliminadas_outliers)

print("\nPorcentaje de filas eliminadas:")
print(round(porcentaje_eliminado_outliers, 2), "%")


# ------------------------------------------------------------
# Comprobamos de nuevo la distribución del precio
# ------------------------------------------------------------

print("\nResumen estadístico del precio después del filtro:\n")
display(df_dashboard["precio"].describe())


# ------------------------------------------------------------
# Revisamos los precios más altos después del filtro
# ------------------------------------------------------------

print("\nAlojamientos con precios más altos después del filtro:\n")

df_dashboard[
    [
        "distrito",
        "nombre",
        "tipo_propiedad",
        "precio",
        "dormitorios",
        "capacidad",
        "ingresos_estimados_365",
        "ocupacion_estimada_365"
    ]
].sort_values(
    by="precio",
    ascending=False
).head(20)
```

## Creación de variables derivadas

Con el dataset ya de valores nulos relevantes y precios extremos, se crean nuevas variables a partir de columnas ya existentes para facilitar el análisis y la visualización en Power BI.

Estas variables no añaden información externa, sino que transforman datos existentes en categorías más comprensibles para una audiencia no técnica.

Las variables creadas serán:

- `tipo_dormitorios`: clasifica las viviendas según número de dormitorios.
- `ingresos_estimados_mensuales`: transforma los ingresos estimados anuales en una media mensual.
- `total_alojamientos_distrito`: mide la competencia turística por distrito.
- `nivel_competencia`: clasifica los distritos según volumen de alojamientos.
- `nivel_precio`: clasifica los alojamientos según su precio relativo dentro del dataset.

```python
# ============================================================
# 15. CREACIÓN DE VARIABLES DERIVADAS
# ============================================================
# Creamos nuevas columnas útiles para el análisis y el dashboard.
# Esta celda debe ejecutarse después de limpiar nulos y outliers.
# ============================================================


# Si la celda se ejecuta más de una vez, eliminamos antes
# las columnas derivadas para evitar duplicados o errores.
columnas_derivadas = [
    "tipo_dormitorios",
    "ingresos_estimados_mensuales",
    "total_alojamientos_distrito",
    "nivel_competencia",
    "nivel_precio"
]

df_dashboard = df_dashboard.drop(
    columns=columnas_derivadas,
    errors="ignore"
)


# ------------------------------------------------------------
# 1. Tipo de dormitorios
# ------------------------------------------------------------

def clasificar_dormitorios(valor):
    if pd.isna(valor):
        return "Sin información"
    elif valor <= 0:
        return "Sin información"
    elif valor == 1:
        return "1 dormitorio"
    elif valor == 2:
        return "2 dormitorios"
    else:
        return "3 o más dormitorios"


df_dashboard["tipo_dormitorios"] = df_dashboard["dormitorios"].apply(clasificar_dormitorios)


# ------------------------------------------------------------
# 2. Ingresos estimados mensuales
# ------------------------------------------------------------
# No es beneficio neto. Solo divide los ingresos estimados anuales entre 12.

df_dashboard["ingresos_estimados_mensuales"] = (
    df_dashboard["ingresos_estimados_365"] / 12
)


# ------------------------------------------------------------
# 3. Total de alojamientos por distrito
# ------------------------------------------------------------

alojamientos_por_distrito = (
    df_dashboard
    .groupby("distrito", as_index=False)
    .agg(total_alojamientos_distrito=("id_alojamiento", "count"))
)


df_dashboard = df_dashboard.merge(
    alojamientos_por_distrito,
    on="distrito",
    how="left"
)


# ------------------------------------------------------------
# 4. Nivel de competencia
# ------------------------------------------------------------

limite_baja_competencia = alojamientos_por_distrito["total_alojamientos_distrito"].quantile(0.33)
limite_alta_competencia = alojamientos_por_distrito["total_alojamientos_distrito"].quantile(0.66)


def clasificar_competencia(total):
    if total <= limite_baja_competencia:
        return "Baja competencia"
    elif total <= limite_alta_competencia:
        return "Media competencia"
    else:
        return "Alta competencia"


df_dashboard["nivel_competencia"] = df_dashboard["total_alojamientos_distrito"].apply(clasificar_competencia)


# ------------------------------------------------------------
# 5. Nivel de precio
# ------------------------------------------------------------

q1_precio = df_dashboard["precio"].quantile(0.25)
q2_precio = df_dashboard["precio"].quantile(0.50)
q3_precio = df_dashboard["precio"].quantile(0.75)


def clasificar_precio(precio):
    if precio <= q1_precio:
        return "Bajo"
    elif precio <= q2_precio:
        return "Medio"
    elif precio <= q3_precio:
        return "Alto"
    else:
        return "Premium"


df_dashboard["nivel_precio"] = df_dashboard["precio"].apply(clasificar_precio)


# ------------------------------------------------------------
# Comprobación rápida
# ------------------------------------------------------------

print("Variables derivadas creadas correctamente.")
print("\nTamaño actual del dataset:")
print(df_dashboard.shape)

df_dashboard[
    [
        "distrito",
        "precio",
        "nivel_precio",
        "dormitorios",
        "tipo_dormitorios",
        "ingresos_estimados_365",
        "ingresos_estimados_mensuales",
        "total_alojamientos_distrito",
        "nivel_competencia"
    ]
].head(10)
```

## Comprobación de variables derivadas

Se revisan las variables derivadas creadas para comprobar que las categorías tienen sentido y que el dataset está preparado para el análisis por distrito.

Además, se genera una tabla resumen por distrito que servirá como primera aproximación para detectar diferencias en precio, ocupación, ingresos estimados y volumen de alojamientos.

```python
# ============================================================
# 16. COMPROBACIÓN DE VARIABLES DERIVADAS
# ============================================================
# Revisamos que las nuevas variables se hayan creado correctamente
# y generamos un primer resumen agregado por distrito.
# ============================================================


# ------------------------------------------------------------
# 1. Comprobación del tamaño del dataset
# ------------------------------------------------------------

print("Tamaño actual del dataset:")
print(df_dashboard.shape)


# ------------------------------------------------------------
# 2. Comprobación de las nuevas variables creadas
# ------------------------------------------------------------

print("\nDistribución por tipo de dormitorios:\n")
print(df_dashboard["tipo_dormitorios"].value_counts())

print("\nDistribución por nivel de precio:\n")
print(df_dashboard["nivel_precio"].value_counts())

print("\nDistribución por nivel de competencia:\n")
print(df_dashboard["nivel_competencia"].value_counts())


# ------------------------------------------------------------
# 3. Comprobación del precio después del filtro de outliers
# ------------------------------------------------------------
# El precio máximo debería ser 1.000 o menos.

print("\nResumen del precio después de la limpieza:\n")
display(df_dashboard["precio"].describe())


# ------------------------------------------------------------
# 4. Resumen agregado por distrito
# ------------------------------------------------------------
# Esta tabla nos ayuda a comparar distritos de Málaga según:
# - volumen de alojamientos
# - precio medio y mediano
# - ocupación media
# - ingresos estimados
# - actividad reciente medida por reviews del último año

resumen_distritos = (
    df_dashboard
    .groupby("distrito")
    .agg(
        total_alojamientos=("id_alojamiento", "count"),
        precio_medio=("precio", "mean"),
        precio_mediano=("precio", "median"),
        ocupacion_media=("ocupacion_estimada_365", "mean"),
        ingresos_medios_anuales=("ingresos_estimados_365", "mean"),
        ingresos_medios_mensuales=("ingresos_estimados_mensuales", "mean"),
        reviews_ultimo_anio=("reviews_ultimo_anio", "sum"),
        reviews_media_ultimo_anio=("reviews_ultimo_anio", "mean")
        
    )
    .reset_index()
)


# ------------------------------------------------------------
# 5. Redondeo de columnas numéricas
# ------------------------------------------------------------

columnas_redondear = [
    "precio_medio",
    "precio_mediano",
    "ocupacion_media",
    "ingresos_medios_anuales",
    "ingresos_medios_mensuales",
    "reviews_media_ultimo_anio"
    
]

resumen_distritos[columnas_redondear] = resumen_distritos[columnas_redondear].round(2)


# ------------------------------------------------------------
# 6. Ordenamos por ingresos medios anuales
# ------------------------------------------------------------
# Esto permite ver qué distritos aparecen con mayores ingresos
# estimados medios, aunque todavía no se deben sacar conclusiones finales.

resumen_distritos = resumen_distritos.sort_values(
    by="ingresos_medios_anuales",
    ascending=False
)


# Mostramos la tabla final
resumen_distritos
```
